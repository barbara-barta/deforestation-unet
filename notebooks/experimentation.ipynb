```python
from pathlib import Path
import os   
import subprocess 
data_dir, local_run = get_data_dir()
if local_run == False:
    from google.colab import drive
```

```python
def get_data_dir():
    if os.path.exists("/content"):
        # running in Google Colab
        local_run = False
        return Path("/content/data"),local_run
    else:
        # running locally
        base_dir = Path(__vsc_ipynb_file__).resolve().parent.parent
        local_run = True
        return base_dir / "data", local_run
```

```python
paths = {
    "rgb": data_dir / "rgb",
    "am4": data_dir / "am4",
    "at4": data_dir / "at4",
}

for path in paths.values():
    Path(path).mkdir(parents=True, exist_ok=True)
    Path(path/"raw").mkdir(parents=True, exist_ok=True)
    Path(path/"processed").mkdir(parents=True, exist_ok=True)
    Path(path/"interrim").mkdir(parents=True, exist_ok=True)
    Path(path/"external").mkdir(parents=True, exist_ok=True)
```

```python

downloads = {
    "rgb": ["https://doi.org/10.5281/zenodo.3233081", "Amazon Forest Dataset.rar"]
}
#  "am4": ["https://doi.org/10.5281/zenodo.4498086","AMAZON.rar"],
#  "at4": ["https://doi.org/10.5281/zenodo.4498086","ATLANTIC FOREST.rar"]}

for dataset, [url, filename] in downloads.items():
    data_empty = not any(Path(paths[dataset]).iterdir())
    if data_empty:
        print(f"Downloading {dataset} data...")
        path = paths[dataset]
        cmd = ["zenodo_get", "-o", str(path), "-g", filename] + [url]
        subprocess.run(cmd, check=True)
        print(f"Dataset {dataset} downloaded to {str(path)}.")
    else:
        print(
            f"{dataset} data already exists at {str(paths[dataset])}, skipping download."
        )
```
